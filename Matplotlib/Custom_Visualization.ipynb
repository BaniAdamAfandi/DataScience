Dengan menggunakan data di bawah ini, plot kan data tersebut kemudian custom sehingga memiliki result seperti yang di harapkan :

note: warna terdiri dari warna Hijau, kuning, Biru, dan Merah

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-10, 10, 100)
cos = np.cos(x)
cos3 = np.cos(x)*3
sin = np.sin(x-90)
sin2 = np.sin(x)*2
```

```python
b = plt.figure(figsize=(9,7))
a = b.add_subplot()
a.plot(x,cos, linestyle='solid', color='red',label='Cosx')
a.plot(x,cos3,marker='', linestyle='--', color='green',label='3 Cosx')
a.plot(x,sin,marker='', linestyle='-.', color='blue',label='Sin(x-90)')
a.plot(x,sin2,marker='x', linestyle=' ', color='yellow',label='2 Sinx')
a.set_xlabel('X Axis Label')
a.set_ylabel('Y Axis Label')
a.set_title('Grafik Cos Sin')
a.legend(loc='center')
plt.show()
```
